(a) On the basis of this dissimilarity matrix, sketch the dendrogram
that results from hierarchically clustering these four observations using complete linkage. Be sure to indicate on the plot the
height at which each fusion occurs, as well as the observations
corresponding to each leaf in the dendrog

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
dissimilarity_matrix = np.array([[0, 0.3, 0.4, 0.7],
                   [0.3, 0, 0.5, 0.8],
                   [0.4, 0.5, 0, 0.45],
                   [0.7, 0.8, 0.45, 0]])
```

```python
#Citation https://numpy.org/doc/stable/reference/generated/numpy.triu_indices.html
r, c = dissimilarity_matrix .shape
data = linkage(dissimilarity_matrix [np.triu_indices(r,1)], method='complete')
```

```python
plt.title('Clustering dendrogram')
plt.xlabel('Observation')
plt.ylabel('Distance')
dendrogram(data)
plt.show()
```

(b)Repeat (a), this time using single linkage clustering

```python
datab = linkage(dissimilarity_matrix [np.triu_indices(r,1)], method='single')
```

```python
plt.title('Clustering dendrogram')
plt.xlabel('Observation')
plt.ylabel('Distance')
dendrogram(datab)
plt.show()
```
