# ISLR 9.7.3

a. We are given n = 7 observations in p = 2 dimensions. For each
observation, there is an associated class label.Sketch the observations


```python
from sklearn import svm
import numpy as np
import matplotlib.pyplot as plt

X = [[3, 2, 4, 1, 2, 4, 4], [4, 2, 4, 4, 1, 3, 1]]
Y = ['red', 'red', 'red', 'red', 'blue', 'blue', 'blue']
```

```python
#Citation: https://amitrajan012.github.io/post/support-vector-machines_part4/
# fit the model
clf = svm.SVC(kernel="linear",C=10000)
clf.fit(np.array(X).T.tolist(), Y)
plt.scatter(X[0], X[1], c=Y,label='Obervations')
# get the separating hyperplane
Coeff = clf.coef_[0] #(n_class * (n_class-1) / 2,n_features
coeff_x1 = -Coeff[0] / Coeff[1]
xx = np.linspace(0, 5)
yy = coeff_x1 * xx - (clf.intercept_[0]) / Coeff[1]

plt.plot(xx, yy, 'g',label='Separating Hyperplane')
plt.grid()
plt.xlabel('X1')
plt.ylabel('X2')
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1],s=200,
             facecolors='none', edgecolors='black',label='Sensitive points')
plt.legend(loc='lower right')
```

b.Sketch the optimal separating hyperplane, and provide the equation for this hyperplane

Result shown above

```python
print('The hyperplane equation is : X2 =',coeff_x1,"* X1 ",- (clf.intercept_[0]) / Coeff[1])
```

c. Describe the classification rule for the maximal margin classifier.
It should be something along the lines of Classify to Red if
0 + 1X1 + 2X2 > 0, and classify to Blue otherwise. Provide
the values for 0, 1, and 2.

 0= - 0.5,  1 = 1, and 2 = -1. When X2-X1-0.5 > 0, it is below the hyperplane, which is blue. When X2-X1-0.5 < 0, it is red.

d. On your sketch, indicate the margin for the maximal margin
hyperplane

```python
# plot the parallels to the separating hyperplane that pass through the
# support vectors
svm_down = clf.support_vectors_[0]
yy_down = coeff_x1 * xx + (svm_down[1] - coeff_x1 * svm_down[0])
svm_up = clf.support_vectors_[-1]
yy_up =coeff_x1 * xx + (svm_up[1] - coeff_x1 * svm_up[0])

plt.scatter(X[0], X[1], c=Y,label='Obervations')
plt.plot(xx, yy, 'g',label='Separating Hyperplane')
plt.plot(xx, yy_down,'k--')
plt.plot(xx,yy_up ,'k--')
plt.grid()
plt.xlabel('X1')
plt.ylabel('X2')
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1],s=200,
             facecolors='none', edgecolors='black',label='Sensitive points')
plt.legend(loc='lower right')
```

g.Sketch a hyperplane that is not the optimal separating hyperplane, and provide the equation for this hyperplane

```python
clf_no = svm.SVC(kernel="linear",C=0.1)
clf_no.fit(np.array(X).T.tolist(), Y)
plt.scatter(X[0], X[1], c=Y,label='Obervations')
# get the separating hyperplane
Coeff_no = clf_no.coef_[0] #(n_class * (n_class-1) / 2,n_features
coeff_x1_no = -Coeff_no[0] / Coeff_no[1]
xx_no = np.linspace(0, 5)
yy_no = coeff_x1_no * xx_no - (clf_no.intercept_[0]) / Coeff_no[1]

plt.plot(xx_no, yy_no, 'g',label='Separating Hyperplane')
plt.grid()
plt.xlabel('X1')
plt.ylabel('X2')
plt.scatter(clf_no.support_vectors_[:, 0], clf_no.support_vectors_[:, 1],s=200,
             facecolors='none', edgecolors='black',label='Sensitive points')
plt.legend(loc='lower right')
plt.title('No-optimal Case')
```
